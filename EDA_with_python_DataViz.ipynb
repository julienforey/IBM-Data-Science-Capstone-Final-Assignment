```python
# install library before
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Collect and plot (exploratory)


First, let's read the SpaceX dataset into a Pandas dataframe and print its summary


```python
df=pd.read_csv('/Users/julienforet/Downloads/IBMFinalproject/dataset_part_2.csv')

df.head(5)
```

```python
sns.catplot(y="PayloadMass", x="FlightNumber", hue="Class", data=df, aspect = 5)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Pay load Mass (kg)",fontsize=20)
plt.show()
```

```python
# Sns for scatter plot
sns.catplot(y="LaunchSite", x="FlightNumber", hue="Class", data=df, aspect = 5)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Launch Site",fontsize=20)
plt.title("Launch Site per flight number",fontsize=20)
plt.show()
```

```python
# Scatter plot again
sns.catplot(y="LaunchSite", x="PayloadMass", hue="Class", data=df, aspect = 5)
plt.xlabel("Pay load Mass (kg)",fontsize=20)
plt.ylabel("Launch Site",fontsize=20)
plt.title("Launch Site per load mass",fontsize=20)
plt.show()
```

```python
# HINT use groupby method on Orbit column and get the mean of Class column
pl = df.groupby('Orbit')['Class'].mean()
ax = pl.plot(kind='bar')
ax.set_xlabel("Orbit")
ax.set_ylabel("Sucess Rate of each Orbit")
ax.set_title("Success per orbite",fontsize=20)
```

```python
# Plot a scatter point chart with x axis to be FlightNumber and y axis to be the Orbit, and hue to be the class value
sns.catplot(y="Orbit", x="FlightNumber", hue="Class", data=df, aspect = 5)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Orbit",fontsize=20)
plt.title("Orbit per flight number",fontsize=20)
plt.show()
```

Theere is a little trend with the flight number increasing


```python
year=[]
def Extract_year(date):
    for i in df["Date"]:
        year.append(i.split("-")[0])
    return year
    
```

```python
# Line chart
df['Year'] = Extract_year(df["Date"])
df_groupby_year = df.groupby("Year",as_index=False)["Class"].mean()
sns.lineplot(data = df_groupby_year, x="Year", y="Class")
plt.xlabel("Year")
plt.title('Success Rate over the years')
plt.ylabel("Success Rate")
plt.show()
```

We can see a function like 1/log(x)
